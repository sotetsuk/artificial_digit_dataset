## Make digits

```python
digits = [np.zeros((dim, dim)) for i in range(10)]
digits[0] = a + b + c + e + f + g + h
digits[1] = a + e + h
digits[2] = a + b + e + d + f + g
digits[3] = a + b + e + d + h + g
digits[4] = a + c + d + e + h
digits[5] = a + b + c + d + g + h
digits[6] = a + b + c + f + d + h + g
digits[7] = a + b + e + h
digits[8] = a + b + c + d + e +f + g + h
digits[9] = a + b + c + d + e + h
```

```python
for i in range(10):
    plt.plot()
    plt.imshow(digits[i], cm.gray, interpolation='nearest')
    plt.show()
```

## Make dataset

bit flip probability

```python
p = 0.3
```

sample size

```python
train_n = 50000
valid_n = 10000
test_n = 10000
```

```python
train_set_x = []
train_set_y = []

for i in range(10):
    for j in range(train_n/10):
        m = np.array(ss.bernoulli(p).rvs((28, 28)), np.bool)
        t = np.copy(digits[i])
        t[m] -= 1.
        t[t == -1.] = 1.
        train_set_x.append(t)
        train_set_y.append(i)
        
train_set_x = np.array(train_set_x, np.int32).reshape(train_n, dim*dim)
train_set_y = np.array(train_set_y, np.int32)
```

```python
valid_set_x = []
valid_set_y = []

for i in range(10):
    for j in range(valid_n/10):
        m = np.array(ss.bernoulli(p).rvs((28, 28)), np.bool)
        t = np.copy(digits[i])
        t[m] -= 1.
        t[t == -1.] = 1.
        valid_set_x.append(t)
        valid_set_y.append(i)
        
valid_set_x = np.array(valid_set_x, np.int32).reshape(valid_n, dim*dim)
valid_set_y = np.array(valid_set_y, np.int32)
```

```python
test_set_x = []
test_set_y = []

for i in range(10):
    for j in range(test_n/10):
        m = np.array(ss.bernoulli(p).rvs((28, 28)), np.bool)
        t = np.copy(digits[i])
        t[m] -= 1.
        t[t == -1.] = 1.
        test_set_x.append(t)
        test_set_y.append(i)
        
test_set_x = np.array(test_set_x, np.int32).reshape(test_n, dim*dim)
test_set_y = np.array(test_set_y, np.int32)
```

# Save

```python
datasets = ((train_set_x, train_set_y), (valid_set_x, valid_set_y), (test_set_x, test_set_y))
```
